Pour chaque sujet de PEPR_Marseille, on calcule dans chaque région, le quantile de son erreur de reconstruction par rapport à tous les autres sujets

```python
import numpy as np
import pandas as pd
import os
import re
```

```python
def compute_quantile(L,a):
    r=np.mean(L <= a)
    return r
```

```python
dir_path= "/neurospin/tmp/fred/models/2025-11-10"
```

```python

def read_error_csv(path):

    if os.path.exists(path):
        df = pd.read_csv(path)
        return df
    return None
```

Pour le sujet ayant la plus grande cross entropy pour le sillon central droit (j=758)

```python
j=758
```

```python

subfolders = sorted([
    f for f in os.listdir(dir_path)
    if os.path.isdir(os.path.join(dir_path, f))
])



# Table to  save the ks distance
df_758 = pd.DataFrame(columns=["region", "quantile"])

for i, folder in enumerate(subfolders):
    print(i)
    regions=re.match(r"(.*?(left|right))", folder).group(1)

    current_path = os.path.join(dir_path, folder)

    PEPR_file= os.path.join(current_path, "PEPR_Marseille", "Reconstruction_error.csv")

    #print(ukb_file)

    # lecture
    error_PEPR = read_error_csv(PEPR_file)
    #print(len(error_ukb))

    L= error_PEPR.iloc[:,1]
    r = compute_quantile(L, L[j])

    df_758.loc[i] = [regions,r ]


print(df_758)
```

```python
df_758.to_csv("Quantile_subject_758")
```
