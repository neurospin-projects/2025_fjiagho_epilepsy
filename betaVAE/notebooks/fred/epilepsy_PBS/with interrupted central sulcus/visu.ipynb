```python
eps=10**(-10)
beta=2/3
```

```python
def compute_cross_entropy(Input, Output): #Input and Output with shape (n,a,b,c)

    loss = -beta*Input * np.log(Output + eps) - (1-beta)*(1 - Input) * np.log(1 - Output + eps)

    loss_per_subject = np.mean(loss, axis=(1, 2, 3))  # shape (n,)
    loss_mean = np.mean(loss_per_subject)
    return loss_mean, loss_per_subject
```

```python
def extremum(L):

    sorted_indices = np.argsort(L)

    indices_min = sorted_indices[:2]
    indices_max = sorted_indices[-2:]

    return indices_min[0], indices_min[1], indices_max[0], indices_max[1]
```

```python
import os
import nibabel as nib
```

```python
def create(L, name, output_dir):
    
    os.makedirs(output_dir, exist_ok=True)

    L = L.astype(np.float32)

    # Construit les chemins de fichiers automatiquement
    npy_path = os.path.join(output_dir, f"{name}.npy")
    nii_path = os.path.join(output_dir, f"{name}.nii.gz")
    minf_path = os.path.join(output_dir, f"{name}.nii.gz.minf")

    # Sauvegardes
    np.save(npy_path, L)
    nib.save(nib.Nifti1Image(L, affine=np.eye(4)), nii_path)
    with open(minf_path, "w") as f:
        f.write(f"{{'shape': {L.shape}, 'name': '{name}'}}")
```
