In this file, the goal to create a .csv file containing, for each region and for each subject in PEPR_Marseille for each subject in PEPR_Marseille (including those who were not selected) the quatiles of the cross entropy, the chamfer distance and the anomaly One-class SVM score among the same region for all the selected subjects as the file containing the reconstruction_error

```python
import numpy as np
import pandas as pd
import os
import re
```

```python
dir_path= "/neurospin/tmp/fred/models/2025-11-10"  
```

```python
selected_index = np.load("/neurospin/dico/fred/Runs/01_betaVAE/Program/2023_jlaval_STSbabies/betaVAE/notebooks/fred/PEPR_Marseille/All the subjects/histogram_non_zero/index_to_save.npy")
```

```python
def compute_quantile(L,a):
    r=np.mean(L <= a)
    return r
```

```python

def read_error_csv(path):

    if os.path.exists(path):
        df = pd.read_csv(path)
        return df
    return None
```

```python
subfolders = sorted([
    f for f in os.listdir(dir_path)
    if os.path.isdir(os.path.join(dir_path, f))
])


for i, folder in enumerate(subfolders):
    print(i)
    regions=re.match(r"(.*?(left|right))", folder).group(1)

    current_path = os.path.join(dir_path, folder, "PEPR_Marseille")

    df_entropy =pd.DataFrame(columns=["ID", "Cross_entropy"])
    df_Chamfer =pd.DataFrame(columns=["ID", "Chamfer distance"])
    df_SVM =pd.DataFrame(columns=["ID", "Anomaly_score"])

    entropy_file=os.path.join(current_path, "Reconstruction_error.csv")
    chamfer_file=os.path.join(current_path, "Chamfer_distance.csv")
    SVM_file=os.path.join(current_path, "Residualisation_Score_SVM.csv")

    entropy=read_error_csv(entropy_file)
    chamfer=read_error_csv(chamfer_file)
    SVM=read_error_csv(SVM_file)

    values_ref_entropy = entropy.iloc[selected_index, -1].to_numpy()
    values_ref_chamfer = chamfer.iloc[selected_index, -1].to_numpy()
    values_ref_SVM = SVM.iloc[selected_index, -1].to_numpy()

    entropy.iloc[:, -1] = entropy.iloc[:, -1].apply(lambda x: compute_quantile(values_ref_entropy,x))
    chamfer.iloc[:, -1] = chamfer.iloc[:, -1].apply(lambda x: compute_quantile(values_ref_chamfer,x))
    SVM.iloc[:, -1] = SVM.iloc[:, -1].apply(lambda x: compute_quantile(values_ref_SVM,x))

    entropy.to_csv(os.path.join(current_path, "Quantiles_Reconstruction_error.csv"))
    chamfer.to_csv(os.path.join(current_path, "Quantiles_Chamfer_distance.csv"))
    SVM.to_csv(os.path.join(current_path, "Quantiles_Residualisation_Score_SVM.csv"))
```
