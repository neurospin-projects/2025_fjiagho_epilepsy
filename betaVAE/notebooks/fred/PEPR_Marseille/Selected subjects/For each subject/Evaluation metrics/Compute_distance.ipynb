In this fiile we suppose that we have a numpy file with shape (a,b,c) representing the mask of a region, and a point A with coordinates (d,e,f). The goal is to compute the distance between the point A and the closest point of the region.

```python
import numpy as np
```

```python
def compute_distance(L, p): #p = np.array([d, e, f])
    coords= np.argwhere(L == 1)  # shape (N, 3)
    distances = np.linalg.norm(coords - p, axis=1)

    return distances.min()
```
